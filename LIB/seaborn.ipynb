```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
# sns.lineplot(df["total_bill"])
# sns.scatterplot(df["total_bill"])
sns.scatterplot(x = "total_bill", y = "tip", data = df)
plt.title("Scatterplot")
plt.show()
```

```python
sns.histplot(df["total_bill"],color="red",edgecolor= "black")
```

```python
sns.barplot(x = "day",y = "total_bill",data = df)
```

```python
sns.kdeplot(df["total_bill"])
```

```python
##sns -- 1 categorical --catplot
#2 numerical -- scatter plot ,lineplot --relational --relplot
```

```python
sns.relplot(x = "total_bill",y = "tip",data = df,
            kind="scatter",
            hue="smoker",
            style="size",
            col="day")
```

```python
#catplot
sns.catplot(x = "day",y = "total_bill",data = df,
            kind="swarm")
```

```python
##countplot 

sns.countplot(x = "day",data = df)
```

```python
##pointplot--shows the mean of the data connected by lines to show the 
##trend
sns.pointplot(x = "day",y = "total_bill",data = df,hue="sex")
```

```python
##boxplot --- spread of data in the form of box
sns.boxplot(x = "day",y = "total_bill",data = df)
```

```python
##violinplot---spread of data in the form of violin
sns.violinplot(x = "day",y = "total_bill",data = df)
```

```python
sns.pairplot(df)
```

```python
sns.histplot(df["tip"],kde=True)
```

```python
sns.boxplot(df["tip"])
```

```python
##outliers handle 
q1 = df["total_bill"].quantile(0.25)
q3 = df["total_bill"].quantile(0.75)
iqr = q3-q1
lower_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
print(lower_limit,upper_limit)
# df[df["total_bill"]<lower_limit]
df[df["total_bill"]>upper_limit]

df = df[(df["total_bill"]>lower_limit) & (df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
```
